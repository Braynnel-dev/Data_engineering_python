# Web request 3

```python
import requests

response = requests.get("https://elpais.com/america")
response.encoding = "utf-8"
```

```python
import bs4

soup = bs4.BeautifulSoup(response.text, "html.parser")
```

```python
links_last_news = soup.select(".headline_md")
news = [new.a["href"] for new in links_last_news]
for new in news:
    print(new)
```

```python
import validators 
news_links = []
for new in news:
    if not validators.url(new):
        news_links.append(f"https://elpais.com/{new}")
for new_link in news_links:
    print(new_link)
```

```python
response2 = requests.get(news_links[3])
response2.encoding = "utf-8"
soup2 = bs4.BeautifulSoup(response2.text, "html.parser")
```

```python
contents = soup2.select(".article_body p")
contents_list = [content for content in contents]
for content in contents_list:
    print(content)
    print("*"*50)
```
