# Data wrangling

```python
# 1. Añadir newspaper_uid al DataFrame

el_universal["newspapes_uid"] = "eluniversal"

el_universal
```

```python
# 2. Obtener el host

from urllib.parse import urlparse

# apply -> nos permite generar transformaciones custom que nosotros requeramos
el_universal["host"] = el_universal["article_links"].apply(lambda article_links: urlparse(article_links).netloc)

el_universal 
```

```python
el_universal["host"].value_counts()
```

# Missing data

```python
# 3. Rellenar datos faltantes

missing_titles_mask = el_universal["title"].isna()
missing_titles_mask.value_counts()
```

```python
# como en el título no tenemos datos faltantes, intentamos con el body
missing_bodies_mask = el_universal["body"].isna()
missing_bodies_mask.value_counts()
```

```python
missing_bodies_mask
```

```python
# en el body vamos a colocar el texto del último pedazo de la url
# [^/]-->queremos que haga match hasta que no encuentre una diagonal adicional
# [^/]+-->que esto puede suceder una o más veces
# ([^/]+)$ --> vamos ir hasta el final de nuestro string 
# (?P<missing_bodies>[^/]+)$ --> colocar un nombre al grupo
el_universal[missing_bodies_mask]["article_links"].str.extract(r"(?P<missing_bodies>[^/]+)$")
```

```python
# applymap nos permite generar un mapa de un valor a otro, es decir una transformación
missing_bodies = (el_universal[missing_bodies_mask]["article_links"]
                     .str.extract(r"(?P<missing_bodies>[^/]+)$")
                     .applymap(lambda body: body.split("-"))
                     .applymap(lambda body_word_list: " ".join(body_word_list))
                 )
missing_bodies
```

```python
el_universal.loc[missing_bodies_mask, "body"] = missing_bodies.loc[:, "missing_bodies"]
el_universal
```

# Additional cleanup

```python
el_universal
```

```python
# 4. Añadir uid a las filas
import hashlib
#hashlib --> normalmente se utiliza para operaciones criptográficas, pero la vamos a utilziar para generar un hash
#            de la URL, de tal manera que tengamos un número único que mapee siempre a esa URL

# axis=0 -->columbas
# axis=1 -->filas
uids = (el_universal
            .apply(lambda row: hashlib.md5(bytes(row["article_links"].encode())), axis=1)
            .apply(lambda hash_object: hash_object.hexdigest())
        )
el_universal["uid"] = uids
# inplace --> le indica que queremos modificar directamente nuestra tabla 
el_universal.set_index("uid", inplace=True)
el_universal
```

```python
# ahora vamos a eliminar los salto de linea presentes en algunos body

stripped_body = (el_universal
                     .apply(lambda row: row["body"], axis=1)
                     .apply(lambda body: list(body))
                     .apply(lambda letters: list(map(lambda letter: letter.replace("\n", ""), letters)))
                     .apply(lambda letters: list(map(lambda letter: letter.replace("\r", ""), letters)))
                     .apply(lambda letters: "".join(letters))
                )
stripped_body
```
